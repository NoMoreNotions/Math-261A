```python
### Load and extract key information from diversityindex
```

```python
div_index = pd.read_csv('school_enrollment_race.csv')
div_index.columns
```

```python
div_index = div_index[div_index["LocationType"] != "State"]
div_index = div_index.drop(columns = ['FIPSCode', 'LocationType', 'TimeFrame', 'DataFormat'])
div_index = div_index[div_index["Data"] != "S"]
div_index
```

```python
div_index['percentage_squared'] = (div_index['Data'].astype(float)/100)**2
diversity_index = div_index.groupby('Location')['percentage_squared'].sum().reset_index()
diversity_index['Diversity_Index'] = diversity_index['percentage_squared'].apply(lambda x: 1 - x)
diversity_index['Location'] = diversity_index['Location'].str.replace(' County', '')
diversity_index = diversity_index.rename(columns={'Location': 'County Name'})
diversity_index = diversity_index.drop(columns = ['percentage_squared'])
```

```python
merged = merged.merge(diversity_index, on = "County Name", how = "outer")
merged = merged[merged["County Code"].notnull()]
merged
```

```python
merged = merged.dropna(subset=['Percentage Standard Exceeded'])
merged = merged.dropna(subset=['Percentage Standard Not Met'])
merged = merged.dropna(subset=['Students Enrolled'])
merged = merged.dropna(subset=['Students Tested'])
merged = merged.dropna(subset=['Students with Scores'])
merged
```
